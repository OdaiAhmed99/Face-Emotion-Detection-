```python
from tensorflow.keras.models import model_from_json
import numpy as np

import cv2
```

```python
with open("model.json", "r") as json_file:
    loaded_model_json = json_file.read()

loaded_model = model_from_json(loaded_model_json)

loaded_model.load_weights('wighet.h5')
```

```python

def resize_image(image):
    im = cv2.resize(image, (48, 48), interpolation=cv2.INTER_LINEAR)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    return im

```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
lable_name={0: 'angry', 1:'fear',2:'happy',3: 'neutral',4:'sad',5: 'surprise'}
#['angry', 'fear', 'happy', 'neutral', 'sad', 'surprise']
```

```python
video_capture = cv2.VideoCapture(0)  

image_count = 0

while True:
    ret, frame = video_capture.read()
    frame = cv2.flip(frame, 1)  # flip (mirrror)

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        face_roi = frame[y:y+h, x:x+w]  # Extract face by bbox 
        im=resize_image(face_roi)
        
        cv2.putText(frame,lable_name[ loaded_model.predict(im[np.newaxis , ...]  ).argmax()], (x, y-5), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 1, cv2.LINE_AA)

        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2) 

    cv2.imshow('Face Emotion ', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()

```
